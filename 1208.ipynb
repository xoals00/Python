# 학생 한명의 점수를 매개변수로 받아서 80점 이상이면 합격/ 아니면 불합격

```python
def test1(a):
    try :
        if int(a) >= 80 :
            return  '합격'
        else :
            return  '불합격'
    except :
        print('숫자만 입력 가능')
```

```python
test1(30)
```

*args = args 라는 변수에 들어오는 모든 인자를 받는다는 뜻

```python
def test3(*args):
    sum_list = []
    print(args)
    for i in args :
        sum_list.append(i+2)
    return sum_list
```

```python
test3(1,2,3,4)
```
